```python
# get dates
def get_dates(dates):
    dates_array = []
    
    for i in range(len(dates)):
        date = str(dates.iloc[i])
        dates_array.append(parser.parse(date))
    return dates_array

euribor_dates = get_dates(euribor['Date'])
covid_dates = get_dates(covid['date'])
```

```python
from sklearn.metrics import mean_squared_error
X = total_cases_per_million
size = int(len(X) * 0.66)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()
for t in range(len(test)):
	model = ARIMA(history, order=(5,1,0))
	model_fit = model.fit(disp=0)
	output = model_fit.forecast()
	yhat = output[0]
	predictions.append(yhat)
	obs = test[t]
	history.append(obs)
	print('predicted=%f, expected=%f' % (yhat, obs))
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
# plot
fig, ax1 = plt.subplots()
fig.suptitle('Predicted values for total cases per million', fontsize=20)

plt.plot(test, label='Real values')
plt.plot(predictions, color = 'red', label='Predicted values')

fig.set_size_inches(18.5, 10.5)
plt.legend()
plt.show()
```

```python
def upsample(X_sl_new):
    #OVERSAMLPING MINORITY -> BALANCING DATA -> FIXING ROC 
    # separate minority and majority classes
    not_dead_sl = X_sl_new[X_sl_new.deceased==0]
    dead_sl = X_sl_new[X_sl_new.deceased==1]

    # upsample minority
    dead_upsampled_sl = resample(dead_sl,
        replace=True, # sample with replacement
        n_samples=len(not_dead_sl), # match number in majority class
        random_state=27) # reproducible results

    # combine majority and upsampled minority
    upsampled = pd.concat([not_dead_sl, dead_upsampled_sl])

    # check new class counts
    return upsampled
```

```python
df_upsampled = upsample(sl)
```

```python
def regressions(sl, cols):
    df = upsample(sl)
    label_df = np.array(df[cols])
    target_df = np.array(df.deceased) 
    X_train_df, X_test_df, y_train_df, y_test_df = train_test_split(label_df, target_df, test_size=0.33, random_state=42)
    #logsitic regression
    lr = LogisticRegression(random_state=0).fit(X_train_df, y_train_df)
    #decision tree
    dt = tree.DecisionTreeClassifier().fit(X_train_df, y_train_df)
    #naive bayes
    nb = GaussianNB().fit(X_train_df, y_train_df)
    #support vector machine
    svmc = svm.SVC().fit(X_train_df, y_train_df)
    #multilayer perceptron
    mlp = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1).fit(X_train_df,y_train_df)
    
    #logistic preds
    y_pred_lr = lr.predict(X_test_df)
    #decision tree preds
    y_pred_dt = dt.predict(X_test_df)
    #support vector preds
    y_pred_svm = svmc.predict(X_test_df)
    #naive bayes preds
    y_pred_nb = nb.predict(X_test_df)
    #multilayer perceptron preds
    y_pred_mlp = mlp.predict(X_test_df)
    
    
    print("LOGISTIC REGRESSION:")
    print("Accuracy Score: ", accuracy_score(y_test_df, y_pred_lr))
    print("ROC: ", roc_auc_score(y_test_df, y_pred_lr))
    print("Error: ", 1-roc_auc_score(y_test_df, y_pred_lr))
    print("Precision/Recall/Fscore/Support", precision_recall_fscore_support(y_test_df, y_pred_lr, average='weighted'))
    
    
    print("DECISION TREE:")
    print("Accuracy Score: ", accuracy_score(y_test_df, y_pred_dt))
    print("ROC: ", roc_auc_score(y_test_df, y_pred_dt))
    print("Error: ", 1-roc_auc_score(y_test_df, y_pred_dt))
    print("Precision/Recall/Fscore/Support", precision_recall_fscore_support(y_test_df, y_pred_dt, average='weighted'))
    
    
    print("MULTI-LAYER PERCEPTRON:")
    print("Accuracy Score: ", accuracy_score(y_test_df, y_pred_mlp))
    print("ROC: ", roc_auc_score(y_test_df, y_pred_mlp))
    print("Error: ", 1-roc_auc_score(y_test_df, y_pred_mlp))
    print("Precision/Recall/Fscore/Support", precision_recall_fscore_support(y_test_df, y_pred_mlp, average='weighted'))
    
    print("SUPPORT VECTOR MACHINE:")
    print("Accuracy Score: ", accuracy_score(y_test_df, y_pred_svm))
    print("ROC: ", roc_auc_score(y_test_df, y_pred_svm))
    print("Error: ", 1-roc_auc_score(y_test_df, y_pred_svm))
    print("Precision/Recall/Fscore/Support", precision_recall_fscore_support(y_test_df, y_pred_svm, average='weighted'))
    
    print("NAIVE BAYES CLASSIFIER:")
    print("Accuracy Score: ", accuracy_score(y_test_df, y_pred_nb))
    print("ROC: ", roc_auc_score(y_test_df, y_pred_nb))
    print("Error: ", 1-roc_auc_score(y_test_df, y_pred_nb))
    print("Precision/Recall/Fscore/Support", precision_recall_fscore_support(y_test_df, y_pred_nb, average='weighted'))
```
